#### adding thresholds to plot ####

```python
setupA_Move2_envelope_baseline = setupA_Move2_enveloped_df[10*SF:15*SF]
setupA_Move2_tkeo_baseline = setupA_Move2_tkeo_df[10*SF:15*SF]

mu_move2_envelope_baseline, sigma_move2_envelope_baseline = setupA_Move2_envelope_baseline.mean(), setupA_Move2_envelope_baseline.std()

mu_move2_tkeo_baseline, sigma_move2_tkeo_baseline = setupA_Move2_tkeo_baseline.mean(), setupA_Move2_tkeo_baseline.std()
```

```python
### ==== envelope ==== ###
k_values_envelope = [15, 15, 10]
envelope_threshs = []
onsets_envelope = []
offsets_envelope = []

for i, channel in enumerate(channels):
    thresh_envelope = mu_move2_envelope_baseline[channel] + k_values_envelope[i] * sigma_move2_envelope_baseline[channel]
    envelope_threshs.append(thresh_envelope)
    envelope_activity = setupA_Move2_enveloped_df[channel] > thresh_envelope
    min_samples = int(0.8 * SF)

    labels, n_lbl = label(envelope_activity)
    valid = np.zeros_like(envelope_activity, dtype=bool)
    
    for lbl in range(1, n_lbl + 1):
        idx = np.where(labels == lbl)[0]
        if idx.size >= min_samples:
            valid[idx] = True
    
    # On- & Offsets
    envelope_onset  = np.where(np.diff(valid.astype(int)) ==  1)[0] + 1
    envelope_offset = np.where(np.diff(valid.astype(int)) == -1)[0] + 1
    
    onsets_envelope.append(envelope_onset)
    offsets_envelope.append(envelope_offset)   
    
### ===== tkeo ==== ###

k_values_tkeo = [8, 8, 10]
tkeo_threshs = []
onsets_tkeo = []
offsets_tkeo = []

for i, channel in enumerate(channels):
    thresh_tkeo = mu_move2_tkeo_baseline[channel] + k_values_tkeo[i] * sigma_move2_tkeo_baseline[channel]
    tkeo_threshs.append(thresh_tkeo)
    tkeo_activity = setupA_Move2_tkeo_df[channel] > thresh_tkeo
    min_samples_tkeo = int(0.8 * SF)

    labels_tkeo, n_lbl_tkeo = label(tkeo_activity)
    valid = np.zeros_like(tkeo_activity, dtype=bool)
    
    for lbl in range(1, n_lbl_tkeo + 1):
        idx = np.where(labels_tkeo == lbl)[0]
        if idx.size >= min_samples_tkeo:
            valid[idx] = True
    
    # On- & Offsets
    onset_tkeo  = np.where(np.diff(valid.astype(int)) ==  1)[0] + 1
    offset_tkeo = np.where(np.diff(valid.astype(int)) == -1)[0] + 1
    
    onsets_tkeo.append(onset_tkeo)
    offsets_tkeo.append(offset_tkeo)   
```

```python
fig, axs = plt.subplots(3, 2, figsize=(12, 8))

channels = ["brachioradialis_L", "deltoideus_L", "SVM_L"]

for i, channel in enumerate(channels):
    axs[i, 0].plot(setupA_Move2_enveloped_df["Sync_Time (s)"], setupA_Move2_enveloped_df[channel])
    axs[i, 0].axhline(envelope_threshs[i], color="r", linestyle="--", label=f"Threshold with k={k_values_envelope[i]}")
    for y in onsets_envelope[i]:
        axs[i, 0].axvline(setupA_Move2_enveloped_df["Sync_Time (s)"][y], ls="--", c="g")
    for z in offsets_envelope[i]:
        axs[i, 0].axvline(setupA_Move2_enveloped_df["Sync_Time (s)"][z], ls="--", c="g")
    axs[i, 0].set_xlabel("Sync_Time (s)")
    axs[i, 0].set_ylabel("Amplitude (µV)" if channel in emg_cols else "Acceleration (g)")
    axs[i, 0].set_xlim(15,20)
    if channel == "SVM_L":
        axs[i, 0].set_title("ACC - SVM_L")
    else:
        axs[i, 0].set_title(f"EMG (envelope) signal for {channel}")
    axs[i, 0].legend()


    axs[i, 1].plot(setupA_Move2_tkeo_df["Sync_Time (s)"], setupA_Move2_tkeo_df[channel])
    axs[i, 1].axhline(tkeo_threshs[i], color="r", linestyle="--", label=f"Threshold with k={k_values_envelope[i]}")
    for y in onsets_tkeo[i]:
        axs[i, 1].axvline(setupA_Move2_tkeo_df["Sync_Time (s)"][y], ls="--", c="g")
    for z in offsets_tkeo[i]:
        axs[i, 1].axvline(setupA_Move2_tkeo_df["Sync_Time (s)"][z], ls="--", c="g")
    axs[i, 1].set_xlabel("Sync_Time (s)")
    axs[i, 1].set_ylabel("Amplitude (µV)" if channel in emg_cols else "Acceleration (g)")
    axs[i, 1].set_xlim(15, 20)
    if channel == "SVM_L":
        axs[i, 1].set_title("ACC - SVM_L")
    else:
        axs[i, 1].set_title(f"EMG (tkeo) signal for {channel}")
    axs[i, 1].legend()

plt.suptitle(f"setupA_Move2 left arm - comparison different preprocessing steps")
plt.tight_layout()
plt.savefig(f"{fig_dir}/plots_compare_envelope_tkeo/setupA_Move2_arm_L_signals_test.png")
#plt.show()
```
